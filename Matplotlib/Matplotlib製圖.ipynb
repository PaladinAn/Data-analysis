### 基本圖形
- 線的說明標籤，plt參數必須增加label=
    - .legend()
- 加入座標點
    - plt參數必須增加marker='.'
        - . * ^
- numpy的生成數列
    - np.arrange(開始, 結束, 間隔)
- 調整圖的尺寸X:Y
    - .figure(figsize=(x, y))
- 圖片輸出
    - plt.savefig('檔名', dpi=)

```python
x = [0, 1, 2, 3]
y = [0, 2, 4, 6]

plt.plot(x, y, label='2x', marker='*', markersize=10)
plt.title('Hello!', fontname='Comic Sans MS' , fontsize=20)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xticks([0, 1, 2, 3])

plt.legend()

plt.show()
```

```python
x = [0, 1, 2, 3]
y = [0, 2, 4, 6]

plt.figure(figsize=(4, 3), dpi= 300)

plt.plot(x, y, label='2x', marker='*', markersize=10)
plt.title('Hello!', fontname='Comic Sans MS' , fontsize=20)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xticks([0, 1, 2, 3])

plt.legend()

# 第二條線
x2 = np.arange(0, 5, 0.5)
y2 = x2**2
plt.plot(x2 , y2, '--', label = 'x^2')
plt.legend()


#plt.grid()
plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# 調整圖的尺寸 (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(8,5), dpi=100)

# Line 1

# Keyword Argument Notation
#plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

# 速打法
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--', label='2x')

## Line 2

# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')

# Plot remainder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')

# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X, Y axis Tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4,])
#plt.yticks([0,2,4,6,8,10])

# Add a legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300)

# Show plot
plt.show()
```

### 長條圖
- 一個很酷的調整
    - 花樣的色塊填充
        - 先把長條圖存到一個變數 bars = plt.bar(labels, values)
        - 利用bars[索引號].set_hatch('填充的字')
     - .pop()，換成索引的東西

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(5,3), dpi=100)

bars = plt.bar(labels, values)

bars[0].set_hatch('|')
bars[1].set_hatch('X')
bars[2].set_hatch('*')


plt.savefig('barchart.png', dpi=300)

plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(5,3), dpi=100)

bars = plt.bar(labels, values)

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.savefig('barchart.png', dpi=300)

plt.show()
```

### 資料出了點問題!!，有lbs這樣無法計算球員體重的分布
- 移除字串
    - .strip()
- 分隔函數
    - 字串str.split(分隔符, 分的次數)
- 取代法(.str指定後再操作)
    - .str.replace('lbs', '')
    - 格式調整 .astype()

```python
fifa['Weight']
```

```python
fifa_s =fifa['Weight']
fifa_ss = fifa_s.str.replace('lbs', '')
```

```python
fifa_ss
```

```python
fifa_sss = fifa_ss.astype(float)
```

```python
fifa_sss
```

```python
light = fifa_sss.loc[fifa_sss < 125].count()
medium = fifa_sss.loc[(fifa_sss >= 125) & (fifa_sss < 150)].count()
noheavy = fifa_sss.loc[(fifa_sss >= 150) & (fifa_sss < 175)].count()
heavy = fifa_sss.loc[(fifa_sss >= 175) & (fifa_sss < 200)].count()
mega = fifa_sss.loc[fifa_sss >= 200].count()

plt.style.use('ggplot')
plt.figure(figsize=(10, 6))
plt.title('Weight Of Players', fontname='Arial', fontsize=20)
plt.pie((light, medium, noheavy, heavy, mega), labels=['< 125 lbs', '125~150 lbs', '150~175 lbs', '175~200 lbs', '>200 lbs'], autopct='%.2f%%', explode=(.3, .1, 0, 0.08, .2))
plt.show()
```

找有用的分析指標
- 用loc
    - 找barcelona 的總分(overall)
    - 找madrid 的總分(overall)

```python
barcelona = fifa.loc[fifa['Club'] == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa['Club'] == 'Real Madrid']['Overall']
revs = fifa.loc[fifa['Club'] == 'New England Revolution']['Overall']


plt.title('Club Comparison')
plt.style.use('default')
bp = plt.boxplot((barcelona, madrid, revs), labels=('Barcelona', 'Madrid', 'NE Revolution'), patch_artist=True)

for box in bp['boxes']:
    # 改外框顏色
    box.set(color='#4286f4', linewidth=2)
    # 箱子填色
    box.set(facecolor = '#e0e0e0' )
    # 箱子填圖案
    #box.set(hatch = '/')
    


plt.xlabel('Club')
plt.ylabel('Gain Point')

plt.show()
```
